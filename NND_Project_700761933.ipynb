```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to compute global histogram of the image
def compute_global_histogram(image):
    """Compute and return the global histogram of an image."""
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    return hist

# Function to apply global context (e.g., histogram equalization)
def enhance_image_with_global_context(image):
    """
    Enhance the image using global context.
    This could be histogram equalization or any other approach
    that uses global context for enhancement.
    """
    # Convert the image to grayscale if it's not already
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    # Calculate the global histogram
    hist = compute_global_histogram(gray_image)

    # Use histogram equalization as a simple form of enhancement
    # Histogram equalization is a global context-based operation
    enhanced_image = cv2.equalizeHist(gray_image)

    return enhanced_image

# Function to display the original and processed images
def display_images(original, processed):
    """Display the original and processed images."""
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Enhanced Image")
    plt.imshow(processed, cmap='gray')
    plt.axis('off')

    plt.show()

# Main function to execute the image processing
def main(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Apply image processing with global context guidance (e.g., histogram equalization)
    enhanced_image = enhance_image_with_global_context(image)

    # Display the original and enhanced images
    display_images(image, enhanced_image)

# Run the program
if __name__ == "__main__":
    # Provide the path to your image
    image_path = "NND project image.jpg"  # Change this to your image path
    main(image_path)
```
